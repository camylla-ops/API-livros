```python
# Link dos dados dos livros
link_acervo = "https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet"

# Carregar os dados
acervo = pd.read_parquet(link_acervo)

```

```python
acervo.head()
```

```python
import pandas as pd

# Lista de anos e partes (2010 a 2020, partes 1 e 2)
anos = range(2010, 2021)  # 2010 até 2020
partes = [1, 2]

# Lista para armazenar todos os dados
todos_dados = []

# Loop para cada arquivo
for ano in anos:
    for parte in partes:
        # Monta o link do arquivo
        url = f"https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{ano}{parte}.csv"

        try:
            # Importa o CSV
            df = pd.read_csv(url, encoding='latin-1', sep=',')

            # Adiciona à lista
            todos_dados.append(df)
            print(f" {ano}-parte{parte} importado | Linhas: {len(df)}")

        except Exception as e:
            print(f" Erro em {ano}-parte{parte}: {str(e)}")

# Junta tudo em um único DataFrame
if todos_dados:
    dados_completos = pd.concat(todos_dados, ignore_index=True)
    print("\n Importação concluída!")
    print(f"Total de registros: {len(dados_completos):,}")
    print(f"Colunas: {dados_completos.columns.tolist()}")
else:
    print("\n Nenhum arquivo foi importado!")
```

```python
dados_completos.head()
```

```python
# Corrigir encoding
dados_completos['tipo_vinculo_usuario'] = dados_completos['tipo_vinculo_usuario'].str.encode('latin-1').str.decode('utf-8')
```

```python
dados_completos.head()
```

```python
# Converter datas
for col in ['data_renovacao', 'data_emprestimo', 'data_devolucao']:
    dados_completos[col] = pd.to_datetime(dados_completos[col], errors='coerce')
```

```python
# Remover duplicatas exatas
emprestimos_clean = dados_completos.drop_duplicates()
print("\n Limpeza inicial concluída - Duplicatas removidas:", len(dados_completos) - len(emprestimos_clean))
```

```python
# Bloco 5: Juntar com dados do acervo
# --------------------------------------------------
dados_completos = pd.merge(
    emprestimos_clean,
    acervo,
    on='codigo_barras',
    how='left'
)
```

```python
# Preencher dados faltantes
dados_completos['biblioteca'] = dados_completos['biblioteca'].fillna('Não identificada')
dados_completos['colecao'] = dados_completos['colecao'].fillna('Não catalogado')

print("\n Merge concluído - Dimensão final do dataset:", dados_completos.shape)
```
